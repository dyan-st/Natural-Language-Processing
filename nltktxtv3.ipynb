```python
import nltk
import string
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
nltk.download('wordnet')
nltk.download('omw-1.4')

f = open("nltktest2.txt", "r", encoding='utf-8')
raw=f.read()

#tokenize string
tokens = nltk.word_tokenize(raw)
text = nltk.Text(tokens)
tktitle2 = list(filter(lambda token: token not in string.punctuation, text))


tktitle=[]
for i in tktitle2:
    c = i.lower()
    tktitle.append(c)

#print(tktitle)

# Step 2: Stopwords
nltk.download("stopwords")
from nltk.corpus import stopwords
stpwrd = nltk.corpus.stopwords.words('english')
#add words you don't want in this list below!
#stop_list = ["n't","'s",".","—","&",",","'","'s",":","'","$","n't","","’","*","“","”"]
stop_list = ["n't","'s",".","—","&",",","'","'s",":","'","$","n't","2023","aug.","says","’","*"]
stpwrd.extend(stop_list)
#print(stop_list)
filtered_list = []
for word in tktitle:
    if word.lower() not in stpwrd:
        filtered_list.append(word)

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemlist = []
for i in filtered_list:
    lemmed=(lemmatizer.lemmatize(i))
    lemlist.append(lemmed)

#print(lemlist)


from nltk.probability import FreqDist
fdist = FreqDist(lemlist)
fdist20=fdist.most_common(20)
print (fdist20)


fd2 = nltk.FreqDist(lemlist)
fd2.plot(10,cumulative=False)
```
